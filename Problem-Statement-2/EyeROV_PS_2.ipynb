```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path ="oculus.jpg"
img = cv2.imread(image_path)
```

### **Converting to Grey Scale**

```python
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (10,10))
plt.imshow(grey)
plt.title("Orginal Image")
plt.show()
```

### **Mean Filter**

```python
mean_filtered = cv2.blur(grey,(5,5))
plt.figure(figsize = (10,10))
plt.imshow(mean_filtered)
plt.title("Mean Filtered Image")
plt.show()
```

### **Median Filter**

```python
median_filtered = cv2.medianBlur(grey,5)
plt.figure(figsize = (10,10))
plt.imshow(median_filtered)
plt.title("Median Filtered Image")
plt.show()
```

### **Histogram Equalization**

```python
hist_equalized = cv2.equalizeHist(grey)
plt.figure(figsize = (10,10))
plt.imshow(hist_equalized)
plt.title("Histogram Equalized Image")
plt.show()
```

### **Unsharp Masking**

```python
gaussian_blur = cv2.GaussianBlur(grey,(9,9),10.0)
unsharp_mask = cv2.addWeighted(grey,1.5,gaussian_blur,-0.5,0)
plt.figure(figsize = (10,10))
plt.imshow(unsharp_mask)
plt.title("Unsharp Masked Image")
plt.show()
```
